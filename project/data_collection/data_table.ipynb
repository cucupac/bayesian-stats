# Represent Collected Data in Table Format

```python
import pandas as pd
from nwi_cities import NWI_CITIES

# Convert the dictionary to a DataFrame
df_cities = pd.DataFrame.from_dict(NWI_CITIES, orient="index")
df_cities.reset_index(inplace=True)
df_cities.rename(columns={"index": "City"}, inplace=True)
df_cities["Coordinates"] = df_cities.apply(
    lambda row: f"({row['lat']}, {row['lng']})", axis=1
)

# Select and order the columns for display
columns = [
    "City",
    "Coordinates",
    "population",
    "fips_code",
    "county",
    "fast_food_percentage",
    "median_income",
]
df_cities = df_cities[columns]

# Display the DataFrame
df_cities
```
